# Character-Region Awareness (CRAFT) for Text Detection

CRAFT localize the individual character regions and link the detected characters to a text instance via text_score_heatmap and link_score_heatmap.

#### <i> 1) Installing craft-text-detector module/library</i>

```python
!pip install craft-text-detector
```

```python
# import CRAFT class
from craft_text_detector import Craft
```

#### <i> 2) Taking in image input from local libraries </i>

```python
# set image path and export folder directory
image = r'C:\Documents\Text Recognition\others\train\18.png' # can be filepath, PIL image or numpy array
output_dir = r'C:\Documents\Text Recognition\others\train'

craft = Craft(output_dir=output_dir, crop_type="poly", cuda=False)
```

<li>Raw sample input image</li>
</ul>

#### <i> 3) Applying CRAFT text detection prediction on input image

```python
# apply craft text detection and export detected regions to output directory
prediction_result = craft.detect_text(image)

# unload models from ram/gpu
craft.unload_craftnet_model()
craft.unload_refinenet_model()
```
